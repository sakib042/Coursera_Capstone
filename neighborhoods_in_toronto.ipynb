### Segmenting and Clustering Neighborhoods in Toronto

```python
import pandas as pd
import numpy as np
```

##### Read table from website

```python
d = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
df = d[0]
df.shape
```

##### Ignore cells with a borough that is Not assigned

```python
df_b = df[df['Borough'] != "Not assigned"].reset_index(drop=True)
df_b.head(10)
```

##### If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough.

```python
df_b.loc[df_b['Neighbourhood'] == "Not assigned", ['Neighbourhood']] = df_b['Borough']
df_b.head(10)
```

##### Grouping same postcode

```python
df_filtered = df_b.groupby(['Postcode','Borough']).Neighbourhood.agg([('Neighbourhood', ', '.join)])
df_filtered
```
